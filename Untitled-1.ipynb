```python
import pycbc
import pycbc.catalog
from prep_functions import *
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Carica i cataloghi
catalog_2 = pycbc.catalog.Catalog(source='gwtc-2.1')
catalog_3 = pycbc.catalog.Catalog(source='gwtc-3')

event_names = list(catalog_2.names) + list(catalog_3.names)

sample_rate = 2048
delta_t = 1.0 / sample_rate
```

```python
X = []
y = []
mass_range = []
masses = []
names = []

for name in event_names:

    event = pycbc.catalog.Merger(name)
    m1 = event.median1d('mass_1_source')
    m2 = event.median1d('mass_2_source')

    if 10 < m1 < 30 and 10 < m2 < 30:

        try:
            strain = event.strain('V1')
        except ValueError:
            print(f"No strain data for {name}, skipping...")
            continue
    
        # Process strain
        strain = pycbc.filter.resample_to_delta_t(strain, delta_t)
        merger_time = event.time
        strain_cropped = strain.time_slice(merger_time - 1.5, merger_time + 1.5)
        data = process_data(strain_cropped)

        M = event.median1d('chirp_mass_source')
        q = m1/m2 if m2 < m1 else m2/m1

        X.append(data)
        y.append((1,M,q))
        masses.append((m1,m2))
        names.append(name)
        mass_range.append('low_mass')
        
    elif 30 < m1 < 50 and 30 < m2 < 50:

        try:
            strain = event.strain('V1')
        except ValueError:
            print(f"No strain data for {name}, skipping...")
            continue
    
        # Process strain
        strain = pycbc.filter.resample_to_delta_t(strain, delta_t)
        merger_time = event.time
        strain_cropped = strain.time_slice(merger_time - 1.5, merger_time + 1.5)
        data = process_data(strain_cropped)

        M = event.median1d('chirp_mass_source')
        q = m1/m2 if m2 < m1 else m2/m1

        X.append(data)
        y.append((1,M,q))
        masses.append((m1,m2))
        names.append(name)
        mass_range.append('mid_mass_I')
```
